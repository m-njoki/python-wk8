```python
# Step 1: Data Collection

# Step 2: Data Loading & Exploration
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("/content/sample_data/owid-covid-data.csv")

# Check the structure
print("Columns in the dataset:")
print(df.columns)

# Preview first 5 rows
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Check for missing values in key columns
key_columns = ['date', 'location', 'total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'total_vaccinations']
missing_values_key_columns = df[key_columns].isnull().sum()

# Display missing values in key columns
print("\nMissing values in key columns:")
print(missing_values_key_columns)

# Step 3: Data Cleaning
# Filter countries of interest (e.g., Kenya, USA, India)
countries_of_interest = ['Kenya', 'USA', 'India']
df_filtered = df[df['location'].isin(countries_of_interest)]

# Create a copy of the filtered DataFrame to avoid modifying a view
df_cleaned = df_filtered.copy()

# Drop rows with missing critical values
df_cleaned.dropna(subset=['date', 'total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'total_vaccinations'], inplace=True)

# Convert the 'date' column to datetime format
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'])

# Fill missing numeric values using the median (alternative to interpolation)
df_cleaned.loc[:, 'total_cases'] = df_cleaned['total_cases'].fillna(df_cleaned['total_cases'].median())
df_cleaned.loc[:, 'total_deaths'] = df_cleaned['total_deaths'].fillna(df_cleaned['total_deaths'].median())
df_cleaned.loc[:, 'new_cases'] = df_cleaned['new_cases'].fillna(df_cleaned['new_cases'].median())
df_cleaned.loc[:, 'new_deaths'] = df_cleaned['new_deaths'].fillna(df_cleaned['new_deaths'].median())
df_cleaned.loc[:, 'total_vaccinations'] = df_cleaned['total_vaccinations'].fillna(df_cleaned['total_vaccinations'].median())

# Check for missing values after cleaning
missing_values_after_cleaning = df_cleaned.isnull().sum()
print("\nMissing values after cleaning:")
print(missing_values_after_cleaning)

# Step 4: Exploratory Data Analysis (EDA)
# Set the style for the plots
sns.set(style="whitegrid")

# Task 1: Plot total cases over time for selected countries
plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    country_data = df_cleaned[df_cleaned['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Task 2: Plot total deaths over time for selected countries
plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    country_data = df_cleaned[df_cleaned['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.title('Total COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Task 3: Compare daily new cases between countries
plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    country_data = df_cleaned[df_cleaned['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.title('Daily New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Task 4: Calculate the death rate: total_deaths / total_cases
df_cleaned.loc[:, 'death_rate'] = df_cleaned['total_deaths'] / df_cleaned['total_cases']

# Plot death rate over time for selected countries
plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    country_data = df_cleaned[df_cleaned['location'] == country]
    plt.plot(country_data['date'], country_data['death_rate'], label=country)

plt.title('COVID-19 Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate (Total Deaths / Total Cases)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Step 5: Visualizing Vaccination Progress
# Task 1: Plot cumulative vaccinations over time for selected countries
plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    country_data = df_cleaned[df_cleaned['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

plt.title('Cumulative COVID-19 Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Task 2: Compare % vaccinated population
# Assuming population data is available in the DataFrame as 'population' column
# If not, you would need to manually input the population for each country
df_cleaned['vaccination_percentage'] = (df_cleaned['total_vaccinations'] / df_cleaned['population']) * 100

# Plot the vaccination percentage over time for selected countries
plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    country_data = df_cleaned[df_cleaned['location'] == country]
    plt.plot(country_data['date'], country_data['vaccination_percentage'], label=country)

plt.title('Percentage of Population Vaccinated Over Time')
plt.xlabel('Date')
plt.ylabel('Vaccination Percentage')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Step 6: Build a Choropleth Map
import plotly.express as px

# Step 1: Prepare the DataFrame with 'iso_code' and 'total_cases'
latest_data = df_cleaned[df_cleaned['date'] == df_cleaned['date'].max()]

# Create a copy of the DataFrame to avoid the SettingWithCopyWarning
choropleth_data = latest_data[['iso_code', 'total_cases', 'location']].copy()

# Remove rows with missing 'iso_code' or 'total_cases' in the copy
choropleth_data.dropna(subset=['iso_code', 'total_cases'], inplace=True)

# Step 2: Create the Choropleth Map
fig = px.choropleth(choropleth_data,
                    locations="iso_code",  # Country iso_code
                    color="total_cases",   # Color intensity based on total cases
                    hover_name="location", # Show country name on hover
                    color_continuous_scale="Viridis", # Choose a color scale
                    labels={'total_cases': 'Total COVID-19 Cases'},
                    title="COVID-19 Total Cases by Country (Latest Data)"
                    )

# Show the map
fig.show()

# Step 7: Insights & Graphs
# Insight 1: India's COVID-19 Surge
india_data = df_cleaned[df_cleaned['location'] == 'India']
india_data['date'] = pd.to_datetime(india_data['date'])
india_data.set_index('date', inplace=True)

# Graph: Total Cases Over Time for India
plt.figure(figsize=(10, 6))
plt.plot(india_data.index, india_data['total_cases'], label='Total Cases', color='blue')
plt.title('Total COVID-19 Cases in India Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Insight 2: Impact of Vaccination on Case Numbers
vaccination_data = df_cleaned.dropna(subset=['total_vaccinations', 'new_cases'])
vaccination_data['date'] = pd.to_datetime(vaccination_data['date'])

# Graph: Vaccinations vs New Cases Over Time for Selected Countries
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='total_vaccinations', data=vaccination_data, label='Total Vaccinations', color='green')
sns.lineplot(x='date', y='new_cases', data=vaccination_data, label='New Cases', color='red')

# Add secondary axis for New Cases
plt.twinx()
sns.lineplot(x='date', y='new_cases', data=vaccination_data, label='New Cases', color='red')

plt.title('Vaccinations vs New Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

# Insight 3: Disparity in Mortality Rates Across Continents
continent_mortality = df_cleaned.groupby('continent')[['total_deaths', 'total_cases']].sum()
continent_mortality['mortality_rate'] = continent_mortality['total_deaths'] / continent_mortality['total_cases']

# Graph: Mortality Rate by Continent
plt.figure(figsize=(8, 6))
continent_mortality['mortality_rate'].sort_values().plot(kind='barh', color='purple')
plt.title('Mortality Rate by Continent')
plt.xlabel('Mortality Rate')
plt.ylabel('Continent')
plt.tight_layout()
plt.show()
```
